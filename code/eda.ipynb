```python
import os
import numpy as np

directory = '../data/train/'

essays_number_of_words = []
for file in os.listdir(directory):
    # print(f"Reading file {directory} / {file}")
    with open(directory + file, 'r') as f:
        file_contents = f.read()
        num_words = len(file_contents.split())
        essays_number_of_words.append(num_words)
        

print(f"Average number of words : {np.mean(essays_number_of_words)}")

    
```
